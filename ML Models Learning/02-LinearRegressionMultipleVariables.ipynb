```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("sample_data/homeprices.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']], df['price'])
```

```python
reg.coef_
reg.intercept_
```

```python
reg.predict([[3000, 3, 4]])
```

```python
hiring_df = pd.read_csv("sample_data/hiring.csv")
hiring_df
```

```python
hiring_df['experience'] = hiring_df['experience'].fillna('zero')
median_test_score = math.floor(hiring_df['test_score(out of 10)'].median())
median_test_score
hiring_df['test_score(out of 10)'] = hiring_df['test_score(out of 10)'].fillna(median_test_score)
hiring_df
```
