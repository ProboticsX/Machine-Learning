```python
df = pd.read_csv("sample_data/carprices.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = df
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
dfle
```

```python
x = dfle[['Car Model', 'Mileage', 'Age(yrs)']]
x
```

```python
y = dfle[['Sell Price($)']]
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("town", OneHotEncoder(), [0])], remainder = 'passthrough') # transform 0th column
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:] #save from dummy variable trap
x
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[0,0,45000,4]])
```

```python
model.score(x,y)
```
