PCA helps to reduce dimensions by choosing the best features from the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
x = df
y = digits.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=30)
```

```python
x.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95) # retain 95% of data
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
x_pca_train, x_pca_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_pca_train, y_train)
model.score(x_pca_test, y_test)
```
